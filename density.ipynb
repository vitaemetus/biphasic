```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = open("dump/73000.xyz").readlines()
xyz = []
for line in file[2:]:
    xyz.append(np.array(line.split(sep=' ')[:4]).astype(float))

xyz = pd.DataFrame(xyz, columns=["type", "x", "y", "z"])
```

```python
xyz1 = xyz.loc[xyz["type"] == 1.0]
xyz2 = xyz.loc[xyz["type"] == 2.0]

data = [xyz, xyz1, xyz2]
colors = ["crimson", "tab:blue", "tab:green"]
labels = ["general", "type 1", "type 2"]

for dataframe, color, label in zip(data, colors, labels):
    values, bins = np.histogram(dataframe["z"], bins=60)
    y = values
    x = [(bins[i] + bins[i+1]) / 2 for i in range(bins.shape[0] - 1)]
    plt.plot(x, y, color=color, label=label)

plt.legend()
```
